```python
#association rule with 10% support and 70% confideence
# With 10% Support
frequent_itemsets = apriori(df, min_support = 0.1, use_colnames= True)
frequent_itemsets.head()
```

```python
# with 70% confidence
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 0.7)
rules.head()
```

```python
# A leverage value of 0 indicates independence. Range will be [-1,1]
# A high conviction value means that the consequent is highly depending on the antecedent and range [0,inf]
rules.sort_values('lift', ascending = False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
rules[['support', 'confidence']].hist()
plt.show()
```

```python
rules[['support', 'confidence', 'lift']].hist(figsize = (15,7))
plt.show()
```

```python
plt.scatter(rules['support'], rules['confidence'])
plt.show()
```

```python
sns.scatterplot('support', 'confidence', data = rules, hue = 'antecedents')
plt.show()
```

```python
# Pre-processing
# As the data is not in transaction formation, and data is discrete we can convert dataset values.
# We are using transaction Encoder, To convert dataset values in 0 and 1 format.
movie1 = pd.get_dummies(movie)
movie1.head()
```
